## Kaggler's age group

```python
def plot_barchart(df, col, order=None, 
                  x_label=None, y1_label=None,
                  y2_label=None, title=None, 
                  figsize=(15,8), 
                  plot_percentage=False,
                  display_table=True, 
                  barh=False):
    
    fig, axs = plt.subplots(figsize=figsize)
    
    if barh: # Horizontal barplot
        g = df[col].value_counts(ascending=True).plot.barh(figsize=figsize, color=colors[0])
        for i,rect in enumerate(g.patches):
            h = rect.get_height()
            w = rect.get_width()
            y = rect.get_y()
            g.annotate(f"{w}", (w, y+h/2), color='k', size=12, ha='left', va='center')
    else: # vertical barplot
        g = sns.countplot(data=df, x=col, order=order, color=colors[0], ax=axs, ec='k')
        for rect in g.patches:
            h = rect.get_height()
            w = rect.get_width()
            x = rect.get_x()
            axs.annotate(f"{h}", (x+w/2, h), color='k', size=12, ha='center', va='bottom')

    g.spines['top'].set_visible(False)
    g.spines['left'].set_visible(False)
    g.spines['right'].set_visible(False)

    g.set_ylabel(y1_label)
    g.set_xlabel(x_label)
    plt.title(title, fontdict={'size':16, 'color':colors[1]})
    
    # create sub dataframe of the column of 'total' & 'percentage'
    sub_df = df[col].value_counts().to_frame()
    sub_df = sub_df.rename(columns={col:'Total'})
    if order:
        sub_df = sub_df.loc[order, :]
    sub_df['Percentage'] = (sub_df['Total']/sub_df['Total'].sum())*100
    sub_df = sub_df.round(2)

    # percentage line plot for vertical barplot
    if plot_percentage:
        ax2 = g.twinx()

        p = g.add_child_axes(ax2)
        p.plot(sub_df['Percentage'].cumsum(), marker='o', color=colors[2])
        p.set_ylabel(y2_label)
        for x,y in enumerate(sub_df['Percentage'].cumsum()):
            p.annotate(f"{y:.2f}%", (x,y+2), ha='center', va='bottom', color=colors[1])
        p.spines['top'].set_visible(False)
        p.spines['left'].set_visible(False)
        p.spines['right'].set_visible(False)

    plt.show()

    if display_table:
        display(sub_df.style.bar(color=colors[0]))
```

```python
age_groups = ['18-21', '22-24', '25-29','30-34',
              '35-39','40-44','45-49','50-54',
              '55-59','60-69','70+']
```

```python
plot_barchart(df, col='Q1', order=age_groups, 
              x_label="Age Group", y1_label="Count", 
              y2_label="Cumulative Relative Frequency", 
              title="Kagglers' Age Group", plot_percentage=True)
```

## Gender Share

```python
gender_df = df.Q2.value_counts().to_frame()
gender_df = gender_df.rename(columns={'Q2':'Total'})
gender_df['Percentage'] = (gender_df.Total / gender_df.Total.sum())*100
gender_df = gender_df.round(2)
# gender_df
```

```python
fig, axs = plt.subplots(figsize=(12,8))
plt.pie(gender_df.Total, autopct='%1.2f%%', 
        startangle=30, 
        explode=(0.1,0.1,0,0.1,0.2))

labels = [index+ ' - '+str(round(pct,2))+'%' for (index, pct) in zip(gender_df.index, gender_df.Percentage)]

plt.legend(labels= labels, loc='upper right', 
           bbox_to_anchor=(1.8,1), fontsize=14,
           facecolor='#fce5e8', edgecolor='k',
           borderpad=0.8,
           title='Gender Shares', title_fontsize=16)
plt.show()
gender_df.style.bar(color=colors[0])
```

## Programming Languages

```python
language_options = ['Python', 'R', 'SQL', 
                    'C', 'C++', 'Java', 
                    'Javascript', 'Julia', 'Swift',
                    'Bash', 'MATLAB', None, 'Other']

lang_df = df.filter(like='Q7')

lang_freq_df = lang_df.describe(include='O').T
lang_freq_df.set_index('top', inplace=True)
lang_freq_df = lang_freq_df[['freq']]
lang_freq_df['Percentage'] = (lang_freq_df.freq / lang_freq_df.freq.sum()) * 100
lang_freq_df = lang_freq_df.round(2)
lang_freq_df.sort_values(by='freq', ascending=False, inplace=True)
```

```python
def plot_pandas_series_bar(series, 
                           x_label=None, 
                           y1_label=None, 
                           y2_label=None,
                           title = None, 
                           figsize=(15,8)):
    
    g = series.plot.bar(figsize=figsize, color=colors[0], rot=0)
    g.set_xlabel(x_label)
    g.set_ylabel(y1_label)
    g.set_title(title, fontdict={'size':16, 'color':colors[1]})

    for rect in g.patches:
        h = rect.get_height()
        w = rect.get_width()
        x = rect.get_x()
        g.annotate(f"{h}", (x+w/2, h), color='k', size=12, ha='center', va='bottom')

    g.spines['top'].set_visible(False)
    g.spines['left'].set_visible(False)
    g.spines['right'].set_visible(False)

    ax2 = g.twinx()

    p = g.add_child_axes(ax2)
    percentage_series = (series / series.sum())*100

    p.plot(percentage_series.cumsum(), marker='o', color=colors[2])
    for x,y in enumerate(percentage_series.cumsum()):
        p.annotate(f"{y:.2f}%", (x,y+2), ha='center', va='bottom', color=colors[1])

    p.spines['top'].set_visible(False)
    p.spines['left'].set_visible(False)
    p.spines['right'].set_visible(False)

    p.set_ylabel(y2_label)
    plt.show()
```

```python
plot_pandas_series_bar(lang_freq_df.freq, 
                       x_label="Programming Language", 
                       y1_label="Count", 
                       y2_label="Relative Frequency", 
                       title="Language choices")
```

```python
for col in lang_df.columns:
    lang_df[col] = lang_df[col].apply(lambda x: 1 if x in language_options else 0)
```

```python
num_lang = lang_df.sum(axis=1).value_counts().to_frame()
num_lang = num_lang.rename(columns={0:'Count'}).loc[range(13),:]

plot_pandas_series_bar(num_lang['Count'], 
                       x_label="Number of Known ProgrammingLanguage", 
                       y1_label="Count", 
                       y2_label="Relative Frequency", 
                       title="Number of known progragramming languages by kagglers")
```
