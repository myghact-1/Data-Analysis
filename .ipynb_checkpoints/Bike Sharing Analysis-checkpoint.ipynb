```python
# importing all useful libraries

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = [15,8]
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.left'] = False
plt.rcParams['axes.spines.right'] = False

from calendar import month_name, day_name
from scipy import stats
import statsmodels.api as sm
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.api import adfuller,hp_filter

import warnings
warnings.filterwarnings('ignore')
```

```python
# Columns Renaming

hourly_df.rename(columns={
    'dteday': 'date',
    'yr': 'year',
    'mnth': 'month',
    'hr': 'hour',
    'holiday': 'is_holiday',
    'workingday': 'is_workingday',
    'hum': 'humidity',
    'cnt': 'total'
},
                 inplace=True)
```

```python
## Grouping columns
time_features = [
    'date', 'season', 'year', 'month', 'hour', 'weekday', 'is_holiday',
    'is_workingday'
]
weather_features = ['weathersit', 'temp', 'atemp', 'humidity', 'windspeed']
target_features = ['casual', 'registered', 'total']
```

```python
## data mapping of categorical columns

months = month_name[1:13]
seasons = ['Winter', 'Spring', 'Summer', 'Fall']
years = ['2011', '2012']
days = day_name[:7]  # ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
days = [days[-1]] + days[:-1]  # ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
weathers = ['Clear', 'Cloudy', 'Light_rain_snow', 'Heavy_rain_snow']

df_copy.season = df_copy.season.map(dict(zip(range(1, 5), seasons)))
df_copy.month = df_copy.month.map(dict(zip(range(1, 13), months)))
df_copy.year = df_copy.year.map(dict(zip(range(2), years)))
df_copy.weekday = df_copy.weekday.map(dict(zip(range(7), days)))
df_copy.weathersit = df_copy.weathersit.map(dict(zip(range(1, 5), weathers)))
```

```python
# changing humidity to percent and windspeed values to minimum (0) and maximum (67)

df_copy.windspeed = df_copy.windspeed * 67
df_copy.humidity = df_copy.humidity * 100


# given data is normalized as (t-t_min)/(t_max-t_min)

# getting original temperatues in celsius

temp_min = -8
temp_max = 39

atemp_min = -16
atemp_max = 50

df_copy['temp'] = df_copy.temp*(temp_max-temp_min) + temp_min
df_copy['atemp'] = df_copy.atemp * (atemp_max-atemp_min) + atemp_min
```

```python
def plot_countplot(df, column_name):
    # draw countplot
    g = sns.countplot(x=column_name, data=df)

    for rect in g.patches:
        h, w = rect.get_height(), rect.get_width()
        x, y = rect.get_x(), rect.get_y()
        plt.annotate(f'{h}', (x + w / 2, y + h),
                     ha='center',
                     va='bottom',
                     fontsize=14,
                     color='k')
    plt.title(f"Countplot for feature - {column_name}", fontdict={'size': 18})
    plt.xlabel(f"{column_name}", fontsize=14)
    plt.ylabel('Count', fontsize=14)
    plt.show()
```

<h3 style="background:#98baff; padding:1rem 2rem"> Hypothesis test for Average Weekdays rides vs Average Weekends Rides</h3>

$ 
    H_0 = \text{Average registered rides over weekdays is equal to average registered rides of weekends} \\
    H_a = \text{Opposite of } H_0
$

```python
weekdays_data = df_copy.registered[~df_copy.weekday.
                                   isin(['Saturday', 'Sunday'])]
weekends_data = df_copy.registered[df_copy.weekday.isin(['Saturday',
                                                         'Sunday'])]

test_result = stats.ttest_ind(weekdays_data, weekends_data)  # two-sided test

print(f"Test Statistic: {test_result[0]:.3f}")
print(f"P-value: {test_result[1]:.3f}")
```

<div style="background:#cbe7da; font-size:1.7rem; line-height:1.5; padding:1rem">
    
p-value is less than 0.05, so we can reject the null hypothesis and conclude that there is statistically significance difference between average registered rides during weekdays and weekends.
</div>

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=weekdays_data,
             kde=True,
             stat='density',
             fill=True,
             element='poly',
             label='Weekdays')
sns.histplot(data=weekends_data,
             kde=True,
             stat='density',
             color='orange',
             fill=True,
             element='poly',
             label='Weekends',
             alpha=0.5)

plt.xlabel('Rides')
plt.legend()
plt.show()
```

- **Hypothesis test for casual users**

$ 
    H_0 = \text{Average casual rides over weekdays is equal to average casual rides of weekends} \\
    H_a = \text{Opposite of } H_0
$

```python
weekdays_data_casual = df_copy.casual[~df_copy.weekday.isin(['Saturday', 'Sunday'])]
weekends_data_casual = df_copy.casual[df_copy.weekday.isin(['Saturday', 'Sunday'])]

test_result = stats.ttest_ind(weekdays_data_casual, weekends_data_casual)

print(f"Test Statistic: {test_result[0]:.3f}")
print(f"P-value: {test_result[1]:.3f}")
```

<div style="background:#cbe7da; font-size:1.7rem; line-height:1.5; padding:1rem">

p-value is less than 0.05, so we can reject the null hypothesis and conclude that there is difference between average casual rides during weekdays and weekends.
    
</div>

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=weekdays_data_casual,
             kde=True,
             stat='density',
             fill=True,
             element='poly',
             label='Weekdays')
sns.histplot(data=weekends_data_casual,
             kde=True,
             stat='density',
             color='orange',
             fill=True,
             element='poly',
             label='Weekends',
             alpha=0.5)

plt.xlabel('Rides')
plt.legend()
plt.show()
```

<div style="background:#cbe7da; font-size:1.7rem; line-height:1.5; padding:1rem">

- ***In conclusion, we can say that there is a statistically significant difference between the number of rides on working days and weekend days for both casual and registered customers.***
</div>

```python
## Hypothesis test for casual users

results = stats.ttest_ind(winter_data.registered, not_winter_data.registered)

print(f"Test Statistic: {results[0]:.3f}")
print(f"P-value: {results[1]:.3f}")
```

<div style="background:#cbe7da; font-size:1.7rem; line-height:1.5; padding:1rem">

**For both casual and registered users we get the p-value less than 0.05. So, there is statistically significant difference in rides between both users during winter and non-winter season.**

```python
def plot_correlation(df, col):

    registered_corr = df['registered'].corr(df[col])
    casual_corr = df['casual'].corr(df[col])

    sns.regplot(data=df_copy,
                x=col,
                y='registered',
                label=f'Registered Rides correlation: {registered_corr:.3f}',
                scatter_kws={'alpha': 0.3})
    sns.regplot(data=df_copy,
                x=col,
                y='casual',
                label=f'Casual Rides correlation: {casual_corr:.3f}',
                scatter_kws={'alpha': 0.3})

    plt.title(f"Correlation Between rides and {col}",
              fontdict={
                  'size': 16,
                  'color': 'green'
              })
    plt.legend()
    plt.show()
```

<h3 style="background:#98baff; padding:1rem 2rem"> Time series analysis </h3>

```python
from statsmodels.tsa.stattools import adfuller
```

```python
daily_rides = df_copy[['registered', 'casual']].resample('D').sum()
daily_rides.head()
```

```python
daily_rides.plot()
plt.ylabel('Rides')
plt.title("Total Daily Rides")
plt.show()
```

```python
## Stationarity Test

result_registered = adfuller(daily_rides.registered)
result_casual = adfuller(daily_rides.casual)

print(f"Rigistered Rides' p-value: {result_registered[1]:.3f}")
print(f"Casual Rides' p-value: {result_casual[1]:.3f}")
```

<div style="background:#cbe7da; font-size:1.7rem; line-height:1.5; padding:1rem">

- For both registered and casual users p-value is greater than 0.5, so we cannot reject the null hypothesis. It means they are non - stationary.
</div>

```python
## Trends and seasonality in data
monthly_rides = daily_rides.resample('MS').sum()


cycle_reg, trend_reg =  hp_filter.hpfilter(monthly_rides.registered, lamb=129600)
cycle_casual, trend_casual =  hp_filter.hpfilter(monthly_rides.casual, lamb=129600)
```

```python
trend_reg.plot();
```

```python
cycle_reg.plot();
```

```python
trend_casual.plot();
```
